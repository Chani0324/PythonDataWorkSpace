# Matplotlib 퀴즈

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

data = {
    '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
    '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
plt.bar(df['영화'], df['평점'])
plt.title('국내 Top8 영화 평점 정보')
plt.xticks(rotation=90)
plt.xlabel('영화')
plt.ylabel('평점')
plt.show()
```

```python
df_group = df.groupby('개봉 연도').mean()
df_group['평점']
```

```python
xvalues = range(len(df_group.index))
plt.plot(xvalues, df_group['평점'], marker='o')

for idx, txt in enumerate(df_group['평점']):
    plt.text(xvalues[idx], df_group['평점'].iloc[idx] + 0.1, txt, ha='center')

plt.xticks(xvalues, df_group.index)
plt.ylim(7, 10)
plt.show()
```

```python
filt_01 = df['평점'] >= 9.0
filt_02 = df['평점'] < 9.0 # df[~filt_01]
```

```python
values = [len(df[filt_01].index), len(df[filt_02].index)]
labels = ['9점 이상', '9점 미만']

plt.pie(values, autopct='%.1f%%', labels=labels)
plt.legend(loc=(1.2, 0.3))
plt.show()
```
